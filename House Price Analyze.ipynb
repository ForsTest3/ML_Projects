```python
import numpy as py # arrays
import pandas as pd # dataframe
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

IMPORT HOUSE PRICE DATA HERE (esc + m)

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
#to use your own data set, it would be this line of code below
# house_price_dataset = pd.read_excel('your_file.xlsx')  # For .xlsx files
```

```python
print(house_price_dataset) #will be messy, simply all the original data and read into arrays
```

```python
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names) 
#This does formatting, making it neater
```

```python
# print first 5 rows of DataFrame
house_price_dataframe.head()
```

```python
# add target column (price) to data frame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
# run again for display
house_price_dataframe.head() #now u can see there is a price
```

```python
# checking number of rows first and columns 2nd in dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# stats measurement of datasets
house_price_dataframe.describe()
```

Checking for correlation in features of the dataset

1. positive correlation
2. negative

```python
correlation = house_price_dataframe.corr() #diagram below shows degree of correlation between each variable
```

```python
# constructing heatmap to understand correlation
plt.figure(figsize=(12,8))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'coolwarm', center = 0)
plt.title("Correlation Matrix Heatmap")
plt.show()
#cbar is color bar, squares for display, 1 decimal point, annotations are the left side variable names, size for everything, blue color
#darker the color, higher correlation, lighter color, lower correlation
```

Splitting data and target 

```python
X = house_price_dataframe.drop(['price'], axis = 1) #dropping column is 1
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting data for TRAINING and TESTING data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2) #20% test, needs same random state value
```

```python
print(X.shape ,X_train.shape, X_test.shape) #80 and 20 percent
```

Model Training: XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training model with X_train, to find patterns in Y_train
model.fit(X_train, Y_train)
```

Evaluation for prediction of training data

```python
# need to get accuracy for prediction on TRAINING data
training_data_prediction = model.predict(X_train) #predicts for X_train
```

```python
print(training_data_prediction)
```

```python
# R square error for evaluation
score_1 = metrics.r2_score(Y_train, training_data_prediction) #predict for y for both errors

# Mean Abs Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R squared error: ", score_1)
print("Mean absolute error: ", score_2)
```

Prediction for Test Data

```python
# acc for prediction of TESTING data
test_data_prediction = model.predict(X_train)
```

```python
# R square error for evaluation
score_1 = metrics.r2_score(Y_train, test_data_prediction)

# Mean Abs Error
score_2 = metrics.mean_absolute_error(Y_train, test_data_prediction)

print("R squared error: ", score_1)
print("Mean absolute error: ", score_2)
```

Visualizing actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices Vs. Predicted Prices")
plt.show()
```
